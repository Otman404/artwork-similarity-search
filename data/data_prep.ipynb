```python
import shutil
import os
IMAGES_DIR = "artwork_data/images/images/"
```

```python
# deleting some unecessary folders
# shutil.rmtree(os.path.join(IMAGES_DIR, "Albrecht_DuΓòá├¬rer"), ignore_errors=True)
shutil.rmtree("artwork_data/resized/")
```

```python
# fixing some issues in folder names
p = os.path.join(IMAGES_DIR,'Albrecht_Durer')
os.rename(os.path.join(IMAGES_DIR, 'Albrecht_Du╠êrer'), p)
for img in os.listdir(p):
    os.rename(os.path.join(p, img), os.path.join(p, img.replace('Albrecht_Du╠êrer', 'Albrecht_Durer')))
```

```python
artists = os.listdir(IMAGES_DIR)
pics_per_artist_list = [os.listdir(os.path.join(IMAGES_DIR, artist)) for artist in artists]
artists_artwork_dict = {artist:pics_per_artist for (artist,pics_per_artist) in zip(artists, pics_per_artist_list)}
print("images per artist: ")
artworks_count = [len(artists_artwork_dict[artist]) for artist in artists_artwork_dict.keys()]
images_per_artist = {artist:image_count for (artist, image_count) in zip(artists, artworks_count)}
images_per_artist
```

```python
# for simplicity sake, let's keep 20 paitings max for each artist

max_pics = 20

for artist in os.listdir(IMAGES_DIR):
    images_path = os.path.join(IMAGES_DIR, artist)
    if len(os.listdir(images_path)) > max_pics:
        for image in os.listdir(images_path)[max_pics:]:
            os.remove(os.path.join(images_path, image))
```

```python
## Put all images in same folder, let's call it "combined/"
combined_folder = os.path.join('artwork_data/images/', 'combined')
if not os.path.exists(combined_folder):
    os.makedirs(combined_folder, exist_ok=True)

for folder in os.listdir(IMAGES_DIR):
    for image in os.listdir(os.path.join(IMAGES_DIR, folder)):
        shutil.copyfile(os.path.join(IMAGES_DIR, folder, image), os.path.join(combined_folder, image))

nbytes = sum(d.stat().st_size for d in os.scandir(combined_folder) if d.is_file())
nbytes
```

```python
# converting this csv to a dict to be used as a payload in our vector search

import csv


csv_filename = 'artwork_data/artists_data.csv'
artists_dict = {}

with open(csv_filename) as f:
    reader = csv.DictReader(f)

    for row in reader:
        # artists_desc_list.append({row['name']: {'years': row['years'], 'genre': row['genre'], 'nationality':row['nationality'], 'bio': row['bio'], 'wikipedia': row['wikipedia'], 'paitnings': row['paintings']}})
        artists_dict[row['name']] = {'years': row['years'], 'genre': row['genre'], 'nationality':row['nationality'], 'bio': row['bio'], 'wikipedia': row['wikipedia'], 'paitnings': row['paintings']}
```
